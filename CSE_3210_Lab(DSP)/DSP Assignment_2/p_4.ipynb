```python
# 1.4
# 4. Implementation of Discrete-Time System Using
# Difference Equation
# Objective: Simulate a discrete-time system governed by a difference equation.

# Task:
# • Consider the following difference equation representing an LTI system:
# y[n] − 0.5y[n − 1] = x[n] + 2x[n − 1]
# • Generate an input signal x[n] = δ[n] (unit impulse).
# • Compute the output y[n] for n = 0 to n = 10 using:
# – Direct iterative computation.
# – scipy.signal.lfilter().
# • Compare the results.

import numpy as np
import scipy.signal as signal
import matplotlib.pyplot as plt

a = [1, -0.5]
b = [1, 2]

x = np.zeros(11)
x[0] = 1

y_direct = np.zeros(11)
for i in range(1, 11):
    y_direct[i] = 0.5 * y_direct[i - 1] + x[i] + 2 * (x[i - 1] if i - 1 >= 0 else 0)

y_lfilter = signal.lfilter(b, a, x)


print("Direct Computation Output:", y_direct)
print("lfilter Output:", y_lfilter)
print('\n')

plt.stem(range(11), y_direct, linefmt='b-', markerfmt='bo', basefmt='r-', label='Direct Iteration')
plt.stem(range(11), y_lfilter, linefmt='g-', markerfmt='go', basefmt='r-', label='lfilter')
plt.xlabel('n')
plt.ylabel('y[n]')
plt.legend()
plt.title('Comparison of Direct Computation and lfilter')
plt.show()

```
