```python
# 1.2
# 2. Auto-correlation and Cross-correlation of Signals
# Objective: Implement auto-correlation and cross-correlation.

# Task:
# • Define two signals:
# x[n] = [3, 1, 0, 2, 5], y[n] = [1, 2, 3, 4, 5]
# • Compute auto-correlation of x[n] using:
# – Direct formula implementation.
# – numpy.correlate() with mode ’full’.
# • Compute cross-correlation between x[n] and y[n].
# • Analyze the correlation results

import numpy as np
import matplotlib.pyplot as plt

def auto_correlation(x):
    N = len(x)
    result = []
    for k in range(-N + 1, N):
        sum_value = 0
        for n in range(N):
            if 0 <= n + k < N:
                sum_value += x[n] * x[n + k]
        result.append(sum_value)
    return result

def cross_correlation(x, y):
    N = len(x)
    M = len(y)
    result = []
    for k in range(-M + 1, N):
        sum_value = 0
        for n in range(N):
            if 0 <= n + k < M:
                sum_value += x[n] * y[n + k]
        result.append(sum_value)
    return result

x = [3, 1, 0, 2, 5]
y = [1, 2, 3, 4, 5]

# Auto-Correlation
auto_corr_direct = auto_correlation(x)
auto_corr_numpy = np.correlate(x, x, mode='full')

# Cross-Correlation
cross_corr_direct = cross_correlation(x, y)
cross_corr_numpy = np.correlate(x, y, mode='full')

# Function to format the output
def format_result(result):
    return [float(f"{val:.1f}") for val in result]

# Printing results
print("Auto-Correlation (Direct):", format_result(auto_corr_direct))
print("Auto-Correlation (numpy.correlate):", format_result(auto_corr_numpy))
print("Cross-Correlation (Direct):", format_result(cross_corr_direct))
print("Cross-Correlation (numpy.correlate):", format_result(cross_corr_numpy))
print('\n')

# Plotting results
fig, axs = plt.subplots(2, 2, figsize=(10, 6))

axs[0, 0].stem(range(len(auto_corr_direct)), format_result(auto_corr_direct))
axs[0, 0].set_title("Auto-Correlation")
axs[0, 0].set_xlabel("Index")
axs[0, 0].set_ylabel("Amplitude")

axs[0, 1].stem(range(len(auto_corr_numpy)), format_result(auto_corr_numpy))
axs[0, 1].set_title("Auto-Correlation (numpy.correlate)")
axs[0, 1].set_xlabel("Index")
axs[0, 1].set_ylabel("Amplitude")

axs[1, 0].stem(range(len(cross_corr_direct)), format_result(cross_corr_direct))
axs[1, 0].set_title("Cross-Correlation")
axs[1, 0].set_xlabel("Index")
axs[1, 0].set_ylabel("Amplitude")

axs[1, 1].stem(range(len(cross_corr_numpy)), format_result(cross_corr_numpy))
axs[1, 1].set_title("Cross-Correlation (numpy.correlate)")
axs[1, 1].set_xlabel("Index")
axs[1, 1].set_ylabel("Amplitude")

plt.tight_layout()
plt.show()
```
