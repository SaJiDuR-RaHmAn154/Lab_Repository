```python
# 1.5 
# 5. Implementation of Discrete Fourier Series (DFS)
# Objective: Compute the Discrete Fourier Series (DFS) coefficients of a periodic
# signal.

# Task:
# • Consider the periodic sequence:
#  x[n] = cos(2πn/N), N = 8
# • Compute the DFS coefficients using:
# – Direct formula implementation.
# – numpy.fft.fft() (for comparison).
# • Plot the magnitude and phase of the DFS coefficients.
# • Discuss the interpretation of the frequency components.

import numpy as np
import matplotlib.pyplot as plt

N = 8
n = np.arange(N)

# signal x[n] = cos(2πn/N)
x_n = np.cos(2 * np.pi * n / N)

X_k = np.zeros(N, dtype=complex)
for k in range(N):
    X_k[k] = np.sum(x_n * np.exp(-1j * 2 * np.pi * k * n / N)) / N


X_k_fft = np.fft.fft(x_n) / N

# plotting magnitude and phase
fig, axs = plt.subplots(2, 1, figsize=(8, 6))
axs[0].stem(range(N), np.abs(X_k), linefmt='b-', markerfmt='bo', basefmt='r-', label='Direct Computation')
axs[0].stem(range(N), np.abs(X_k_fft), linefmt='g-', markerfmt='go', basefmt='r-', label='FFT')
axs[0].set_title("Magnitude of DFS Coefficients")
axs[0].set_xlabel("k")
axs[0].set_ylabel("|X[k]|")
axs[0].legend()

axs[1].stem(range(N), np.angle(X_k), linefmt='b-', markerfmt='bo', basefmt='r-', label='Direct Computation')
axs[1].stem(range(N), np.angle(X_k_fft), linefmt='g-', markerfmt='go', basefmt='r-', label='FFT')
axs[1].set_title("Phase of DFS Coefficients")
axs[1].set_xlabel("k")
axs[1].set_ylabel("∠X[k] (radians)")
axs[1].legend()

plt.tight_layout()
plt.show()
```
