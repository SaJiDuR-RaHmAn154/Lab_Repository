```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
```

```python
# Load Iris dataset
iris_df = pd.read_csv('iris.csv')

print(f"Dataset dimensions: {iris_df.shape[0]} rows × {iris_df.shape[1]} columns")
print("\nFirst 5 rows:")
print(iris_df.head())
print("\nData types:")
print(iris_df.dtypes)
```

```python
# features and labels
features = iris_df.drop('variety', axis=1)
labels = iris_df['variety']

print("Features (independent variables):", features.columns.tolist())
print("Label (dependent variable): variety")
print("Number of labels:", labels.nunique())
print("Labels:", labels.unique())
```

```python
# Iris Dataset Analysis
print("=== IRIS DATASET ANALYSIS ===")
print(f"Dataset dimensions: {iris_df.shape[0]} rows × {iris_df.shape[1]} columns")
print(f"Data types: {iris_df.dtypes.to_dict()}")
print(f"Missing values: {iris_df.isnull().sum().sum()}")

# Identify and show labels
print(f"\nLabel Analysis:")
print(f"Number of labels: {labels.nunique()}")
print(f"Label categories: {labels.unique()}")
print(f"Label distribution:")
label_counts = labels.value_counts()
print(label_counts)

# Simple visualization of labels
plt.figure(figsize=(10, 4))

# Label distribution bar chart
plt.subplot(1, 2, 1)
label_counts.plot(kind='bar', color=['skyblue', 'lightcoral', 'lightgreen'])
plt.title('Distribution of Iris Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.xticks(rotation=45)

# Simple scatter plot showing one feature relationship
plt.subplot(1, 2, 2)
colors = ['red', 'blue', 'green']
for i, species in enumerate(labels.unique()):
    species_data = iris_df[iris_df['variety'] == species]
    plt.scatter(species_data['sepal.length'], species_data['sepal.width'], 
                c=colors[i], label=species, alpha=0.7)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Sepal Length vs Width by Species')
plt.legend()

plt.tight_layout()
plt.show()

# Problem type identification
print(f"\n=== PROBLEM TYPE ===")
print(f"Target variable: Categorical (Species)")
print(f"Problem type: MULTICLASS CLASSIFICATION")
print(f"Reason: 3 distinct categories to predict")
```

```python
X_temp, X_test, y_temp, y_test = train_test_split(features, labels, test_size=0.2, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=0.25, random_state=42)

print(f"Training set: {len(X_train)} samples (60%)")
print(f"Validation set: {len(X_val)} samples (20%)")
print(f"Test set: {len(X_test)} samples (20%)")
```

```python
# Online Retail dataset
retail_df = pd.read_excel('Online Retail.xlsx')

print("Dataset shape:", retail_df.shape)
print("First 5 rows:")
print(retail_df.head())
print("\nData types:")
print(retail_df.dtypes)
```

```python
# Online Retail Dataset Analysis
print(f"Dataset dimensions: {retail_df.shape[0]} rows × {retail_df.shape[1]} columns")
print(f"Missing values: {retail_df.isnull().sum().sum()}")

# Cleaning data for analysis
retail_clean = retail_df.dropna(subset=['CustomerID'])
retail_clean['TotalPrice'] = retail_clean['Quantity'] * retail_clean['UnitPrice']

print(f"\nAfter cleaning: {retail_clean.shape[0]} rows")
print(f"Unique customers: {retail_clean['CustomerID'].nunique()}")
print(f"Unique products: {retail_clean['StockCode'].nunique()}")

# For regression analysis - predicting TotalPrice
regression_features = retail_clean[['Quantity', 'UnitPrice']]
regression_labels = retail_clean['TotalPrice']

print(f"\nLabel Analysis:")
print(f"Label: TotalPrice (continuous variable)")
print(f"Price range: ${regression_labels.min():.2f} to ${regression_labels.max():.2f}")
print(f"Average: ${regression_labels.mean():.2f}")

# Simple visualizations
plt.figure(figsize=(10, 4))

# Monthly sales
plt.subplot(1, 2, 1)
retail_clean['Month'] = retail_clean['InvoiceDate'].dt.month
monthly_sales = retail_clean.groupby('Month')['TotalPrice'].sum()
monthly_sales.plot(kind='bar', color='skyblue')
plt.title('Monthly Sales Distribution')
plt.xlabel('Month')
plt.ylabel('Total Sales')

# Price vs Quantity relationship
plt.subplot(1, 2, 2)
sample_data = retail_clean.sample(1000)
plt.scatter(sample_data['Quantity'], sample_data['UnitPrice'], alpha=0.5, s=10)
plt.title('Quantity vs Unit Price')
plt.xlabel('Quantity')
plt.ylabel('Unit Price')

plt.tight_layout()
plt.show()

```

```python
retail_regression = retail_clean.copy()
regression_features = retail_regression[['Quantity', 'UnitPrice']]
regression_labels = retail_regression['TotalPrice']

print("Problem Type: REGRESSION (Predicting TotalPrice)")
print(f"Features: {regression_features.columns.tolist()}")
print(f"Target: TotalPrice")
print(f"Dataset size: {len(regression_features)} samples")


```

```python
# Problem type identification
print(f"\n=== PROBLEM TYPE ===")
print(f"Target variable: Continuous (TotalPrice)")
print(f"Problem type: REGRESSION")
print(f"Reason: Predicting continuous numerical values")
```

```python

# Split the data
X_temp, X_test, y_temp, y_test = train_test_split(regression_features, regression_labels, test_size=0.2, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=0.25, random_state=42)

print(f"Training set: {len(X_train)} samples (60%)")
print(f"Validation set: {len(X_val)} samples (20%)")
print(f"Test set: {len(X_test)} samples (20%)")
```

```python
# Housing dataset
housing_df = pd.read_csv('Housing.csv')

print("Dataset shape:", housing_df.shape)
print("First 5 rows:")
print(housing_df.head())
print("\nData types:")
print(housing_df.dtypes)
```

```python
# Housing Dataset Analysis
print(f"Dataset dimensions: {housing_df.shape[0]} rows × {housing_df.shape[1]} columns")
print(f"Missing values: {housing_df.isnull().sum().sum()}")

# Label analysis
house_prices = housing_df['price']
print(f"\nLabel Analysis:")
print(f"Label: price (continuous variable)")
print(f"Price range: ${house_prices.min():,} - ${house_prices.max():,}")
print(f"Average price: ${house_prices.mean():,.0f}")

# Features analysis
numerical_features = ['area', 'bedrooms', 'bathrooms', 'stories', 'parking']
print(f"\nFeatures: {numerical_features}")

# Simple visualizations
plt.figure(figsize=(10, 4))

# Price distribution
plt.subplot(1, 2, 1)
plt.hist(house_prices/1000000, bins=20, alpha=0.7, color='skyblue', edgecolor='black')
plt.title('Price Distribution (Millions)')
plt.xlabel('Price (Million $)')
plt.ylabel('Frequency')

# Area vs Price relationship
plt.subplot(1, 2, 2)
plt.scatter(housing_df['area'], house_prices/1000000, alpha=0.6, color='green')
plt.title('Area vs Price')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price (Million $)')

plt.tight_layout()
plt.show()

print(f"\n=== PROBLEM TYPE ===")
print(f"Target variable: Continuous (Price)")
print(f"Problem type: REGRESSION")
print(f"Reason: Predicting continuous numerical values")
```

```python
# Features (independent variables)
numerical_features = ['area', 'bedrooms', 'bathrooms', 'stories', 'parking']
features = housing_df[numerical_features]

# Labels (dependent variable)
labels = housing_df['price']

print("Features (independent variables):", features.columns.tolist())
print("Label (dependent variable): price")
print(f"Number of features: {len(features.columns)}")
print(f"Dataset size: {len(features)} samples")
```

```python
X_temp, X_test, y_temp, y_test = train_test_split(features, labels, test_size=0.2, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=0.25, random_state=42)

print(f"Training set: {len(X_train)} samples (60%)")
print(f"Validation set: {len(X_val)} samples (20%)")
print(f"Test set: {len(X_test)} samples (20%)")
```
