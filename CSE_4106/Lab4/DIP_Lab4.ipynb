```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# Thresholding
img = cv2.imread("lion.jpg", 0)

thresholded_img = np.where((img >= 100) & (img < 200), 200, 50)

plt.subplot(1,2,1), plt.imshow(img, cmap="gray"), plt.title("Original")
plt.axis('off')
plt.subplot(1,2,2), plt.imshow(thresholded_img, cmap="gray"), plt.title("Thresholding")
plt.axis('off')
plt.show()
```

```python
# Image Negative
neg_img = 255 - img

plt.subplot(1,2,1), plt.imshow(img, cmap="gray"), plt.title("Original")
plt.axis('off')
plt.subplot(1,2,2), plt.imshow(neg_img, cmap="gray"), plt.title("Negative")
plt.axis('off')
plt.show()
```

```python
# Log Transformation
img = cv2.imread("lion.jpg", 0).astype(np.float32)

c = 255 / np.log(1 + np.max(img))
log_transformed_img = c * np.log(1 + img)
log_transformed_img = np.clip(log_transformed_img, 0, 255).astype(np.uint8)

plt.subplot(1,2,1), plt.imshow(img, cmap="gray"), plt.title("Original")
plt.axis('off')
plt.subplot(1,2,2), plt.imshow(log_transformed_img, cmap="gray"), plt.title("Log Transform")
plt.axis('off')
plt.show()
```

```python
# Gamma Transformation (gamma < 1)
img = cv2.imread("lion.jpg", 0)
gamma = 0.5
C = 255
r = img/255

gamma_img = C * r**gamma

plt.subplot(1,2,1), plt.imshow(img, cmap="gray"), plt.title("Original")
plt.axis('off')
plt.subplot(1,2,2), plt.imshow(gamma_img, cmap="gray"), plt.title("Gamma Transform (γ<1)")
plt.axis('off')
plt.show()
```

```python
# Gamma Transformation (gamma > 1)
gamma = 2
C = 255
r = img/255

gamma_img = C * r**gamma

plt.subplot(1,2,1), plt.imshow(img, cmap="gray"), plt.title("Original")
plt.axis('off')
plt.subplot(1,2,2), plt.imshow(gamma_img, cmap="gray"), plt.title("Gamma Transform (γ>1)")
plt.axis('off')
plt.show()
```

```python
# Contrast Stretching
min_val, max_val = np.min(img), np.max(img)
contrast_stretched = 255 * ((img - min_val) / (max_val - min_val))

plt.subplot(1,2,1), plt.imshow(img, cmap="gray"), plt.title("Original")
plt.axis('off')
plt.subplot(1,2,2), plt.imshow(contrast_stretched, cmap="gray"), plt.title("Contrast Stretched")
plt.axis('off')
plt.show()
```

```python
# Bit plane slicing
fig, axes = plt.subplots(2, 4, figsize=(12, 6))
axes = axes.flatten()

for i in range(8):
    axes[i].imshow(img & (1<<i), cmap='gray')
    axes[i].set_title(f"Bit plane: {i+1}")
    axes[i].axis('off')

plt.tight_layout()
plt.show()
```
